```python
import numpy as np

width = 5

# We can use True and False to represent a █ or a " "
pat0 = np.array([True,True,True], dtype=bool)
pat1 = np.array([False,True,True], dtype=bool)
pat2 = np.array([True,False,True], dtype=bool)
pat3 = np.array([False,False,True], dtype=bool)
pat4 = np.array([True,True,False], dtype=bool)
pat5 = np.array([False,True,False], dtype=bool)
pat6 = np.array([True,False,False], dtype=bool)
pat7 = np.array([False,False,False], dtype=bool)

rows = 5
cols = 5

grid=np.zeros((rows,cols), dtype=bool)

row0 = np.array([True,False,False,True,False], dtype =  bool)
grid[0] = row0

for r in range((rows - 1)):  # Range is rows -1 as we already have row[0]
    # Copy the current working row to it's own variable for 
    # ease of use
    print(f"\n\n **** New Row ****", r)
    this_row = grid[r]
    print(this_row)
    
    # Initialize the nxt_row
    nxt_row = np.zeros(cols, dtype=bool)
    
    idx = len(this_row) - 1
    for i in range(cols):
        # Implement periodic wrap to cope with ends of array
        # Had to look up how to handle "ends" of a row, there are several
        # methods
        # I create the appropriate 3 element array to test
        if i == 0:
            tst_row = this_row[[idx, i, (i+1)]]
        elif i == idx:
            tst_row = this_row[[(i-1),i,0]]
        else:
            tst_row = this_row[[(i-1), i, (i+1)]]
        print(tst_row)
        
        # Check against patterns
        if np.array_equal(tst_row, pat0):
            print("Pat0")
            nxt_row[i]= False
        if np.array_equal(tst_row, pat1):
            print("Pat1")
            nxt_row[i] = True
        if np.array_equal(tst_row, pat2):
            print("Pat2")
            nxt_row[i] = False
        if np.array_equal(tst_row, pat3):
            print("Pat3")
            nxt_row[i] = True
        if np.array_equal(tst_row, pat4):
            print("Pat4")
            nxt_row[i] = False
        if np.array_equal(tst_row, pat5):
            print("Pat5")
            nxt_row[i] = False
        if np.array_equal(tst_row, pat6):
            print("Pat6")
            nxt_row[i] = True
        if np.array_equal(tst_row, pat7):
            print("Pat7")
            nxt_row[i] = True

        # Assign nxt_row[i] to next row in grid
        grid[(r+1)] = nxt_row

```
