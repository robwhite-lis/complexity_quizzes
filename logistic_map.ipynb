```python
# start with the basic imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_n = 10

n = 1

x = np.zeros(max_n, dtype=np.float64)

# Need to set x to some non-zero value to get the sequence going

x[0] = 0.5

# Set up plot
plt.figure(num=0,dpi=120) 
plt.title("Logistic Map Sequence")
plt.xlabel("Index")
plt.ylabel("Value ")

# Loop through lambda values
for lbd in [0.5, 1.0, 2.0]:
    n = 1
    while n < max_n:
        x[n] = lbd * x[n-1]*(1-x[n-1])
        n += 1

    xlist = list(range(max_n))
    ylist = x.tolist()

    plt.plot(xlist,ylist, label=f"lambda = {lbd}")



plt.legend()
plt.show()
```
