```python
import numpy as np
import matplotlib.pyplot as plt 


r = 0.5
max_n = 20
x0_vector = [0.1,0.3,0.5,0.7,0.9]
n = np.arange(1,max_n+1)
xn = np.zeros(max_n)

for x0 in x0_vector:
    xn[0] = x0

    for i in np.arange(0,max_n-1):
        xn[i+1] = r * xn[i] * (1-xn[i])

    plt.plot(n, xn, 'o:', markersize=5, label='x_0='+str(x0)) 


plt.title('Number of bacteria vs time for r='+str(r))
plt.ylabel('x_n (proportion of bacteria)')
plt.xlabel('n (hours)')
plt.legend()
```
